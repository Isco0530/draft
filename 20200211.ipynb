```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
pd.set_option('display.float_format',lambda x : '%.4f' % x)
```

```python
class Total():
    def __init__(self,end_of_quarter,end_of_last_quarter):
        self.end_of_quarter=end_of_quarter
        self.end_of_last_quarter=end_of_last_quarter
        
    def G0102(self):
        name_of_excel=str(self.end_of_quarter)+'-G0102-161-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[4:14,1:3]
        data1.set_index('Unnamed: 1',inplace=True)
        
        name_of_excel1='G0102huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(8).reshape((1,8)),columns=huizong.columns) 
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,0]
        new.iloc[0,2]=data1.iloc[3,0]
        new.iloc[0,3]=data1.iloc[4,0]
        new.iloc[0,4]=data1.iloc[9,0]
        new.iloc[0,5]=new.iloc[0,2]/new.iloc[0,1]
        new.iloc[0,6]=new.iloc[0,3]/new.iloc[0,1]
        new.iloc[0,7]=new.iloc[0,4]/new.iloc[0,1]
        
        name_of_excel2='G0102huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat.to_excel(name_of_excel2,index=False)
        
    def G2200(self):
        name_of_excel=str(self.end_of_quarter)+'-G2200-091-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[24:29,1:5]

        name_of_excel1='G2200huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(3).reshape((1,3)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,3]
        new.iloc[0,2]=data1.iloc[4,3]

        name_of_excel2='G2200huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat1=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat1.to_excel(name_of_excel2,index=False)
        
    def G4400(self):
        name_of_excel=str(self.end_of_quarter)+'-G4400-151-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[4:28,1:3]
        
        name_of_excel1='G4400huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(7).reshape((1,7)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,1]
        new.iloc[0,2]=data1.iloc[1,1]
        new.iloc[0,3]=data1.iloc[5,1]
        new.iloc[0,4]=data1.iloc[13,1]
        new.iloc[0,5]=data1.iloc[18,1]
        new.iloc[0,6]=data1.iloc[23,1]

        name_of_excel2='G4400huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat2=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat2.to_excel(name_of_excel2,index=False)
        
    def G4000(self):
        name_of_excel=str(self.end_of_quarter)+'-G4000-141-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[3:36,1:3]
        
        name_of_excel1='G4000huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(15).reshape((1,15)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,1]
        new.iloc[0,2]=data1.iloc[1,1]
        new.iloc[0,3]=data1.iloc[2,1]
        new.iloc[0,4]=data1.iloc[3,1]
        new.iloc[0,5]=data1.iloc[5,1]
        new.iloc[0,6]=data1.iloc[8,1]
        new.iloc[0,7]=data1.iloc[11,1]
        new.iloc[0,8]=data1.iloc[17,1]
        new.iloc[0,9]=data1.iloc[20,1]
        new.iloc[0,10]=data1.iloc[23,1]
        new.iloc[0,11]=data1.iloc[29,1]
        new.iloc[0,12]=data1.iloc[30,1]
        new.iloc[0,13]=data1.iloc[31,1]
        new.iloc[0,14]=data1.iloc[32,1]

        name_of_excel2='G4000huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat3=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat3.to_excel(name_of_excel2,index=False)
        
    def G0109(self):
        name_of_excel=str(self.end_of_quarter)+'-G0109-161-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[3:15,1:5]
        
        name_of_excel1='G0109huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(2).reshape((1,2)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[10,3]

        name_of_excel2='G0109huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat4=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat4.to_excel(name_of_excel2,index=False)
        
    def G4A01a(self):
        name_of_excel=str(self.end_of_quarter)+'-G4A01a-131-境内汇总数据-人民币'+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[3:18,1:3]
        
        name_of_excel1='G4A01ahuizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(8).reshape((1,8)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,1]
        new.iloc[0,2]=data1.iloc[9,1]
        new.iloc[0,3]=data1.iloc[12,1]
        new.iloc[0,4]=data1.iloc[13,1]
        new.iloc[0,5]=data1.iloc[14,1]
        new.iloc[0,6]=data1.iloc[0,1]/(data1.iloc[12,1]+data1.iloc[13,1]+data1.iloc[14,1])
        new.iloc[0,7]=data1.iloc[0,1]/data1.iloc[9,1]

        name_of_excel2='G4A01ahuizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat5=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat5.to_excel(name_of_excel2,index=False)
        
    def G2501(self):
        name_of_excel='G2501_流动性覆盖率 '+str(self.end_of_quarter)+'.xls'
        data=pd.read_excel(name_of_excel)
        data1=data.iloc[166:167,1:3]
        
        name_of_excel1='G2501huizong'+str(self.end_of_last_quarter)+'.xlsx'
        huizong=pd.read_excel(name_of_excel1)
        new=pd.DataFrame(np.arange(2).reshape((1,2)),columns=huizong.columns)
        new.iloc[0,0]=self.end_of_quarter
        new.iloc[0,1]=data1.iloc[0,1]

        name_of_excel2='G2501huizong'+str(self.end_of_quarter)+'.xlsx'
        self.wechat6=pd.concat([huizong,new],axis=0,ignore_index=True)
        self.wechat6.to_excel(name_of_excel2,index=False)
        
    def total(self):
        self.G0102()
        self.G2200()
        self.G4400()
        self.G4000()
        self.G0109()
        self.G4A01a()
        self.G2501()
        name_of_excel='total'+str(self.end_of_quarter)+'.xlsx'
        total=self.wechat.merge(self.wechat1,on='日期').merge(self.wechat2,on='日期').\
        merge(self.wechat3,on='日期').merge(self.wechat4,on='日期').merge(self.wechat5,on='日期').\
        merge(self.wechat6,on='日期')
        total.to_excel(name_of_excel,index=False)
        print(total)
```

```python
a=Total('20191231','20190930')
a.total()
```

每个季度更新时，先把7张表（G0102,G2200,G4400,G4000,G0109,G4A01a,G2501）放入文件夹
再修改两个日期即可，第一个日期为当季季末，第二个日期为上季季末，
举例20190331时点，a=Total('20190331','20181231')

若表名，表内内容发生修改 或 需要新增表 则需要先修改代码 再按上述步骤操作
